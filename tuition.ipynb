```python
income2 = income.rename(columns = {"name_x":"Instnm"})
```

```python
SEM_Merge = df.merge(income2, on = 'Instnm', how = 'inner')
```

```python
SEM_Merge.shape
```

```python
SEM_Table = SEM_Merge[['Unitid','City','Locale','Instnm','Applcn','Tuition2','Tuition3','Tuition6','Tuition7',\
    'Hloffer','Iclevel','Prgmofr','Upgrntn','Admssn','Applfeeg','Enrlt','Rmbrdamt','Stufacr',\
'F3E07','Enrtot','public', 'tier', 'tier_name','k_median_x','k_mean','k_rank']]
```

```python
df2 = SEM_Table
```

```python
df2.count()
```

get dummy for school tier, ranking 1,2,3

```python
df3 = df2
df3['Tier']=0
df3.loc[(df2['tier']==5) | (df2['tier']==6),'Tier']=1
df3.loc[(df2['tier']==3) | (df2['tier']==4),'Tier']=2
df3.loc[(df2['tier']==1) | (df2['tier']==2),'Tier']=3
df3['Tier'].astype('float').value_counts()
```

get dummy for city, city = 1, otherwise = 0

```python
df3['City_Suburb']=0
df3.loc[(df2['Locale']==11) | (df2['Locale']==12),'City_Suburb']=1
df3['City_Suburb'].value_counts()
```

```python
df3['Appratio']=df3['Applcn']/df3['Enrtot']
df3['Expratio']=df3['F3E07']/df3['Enrtot']
df3['gap']=df3['Tuition3']-df3['Expratio']
```

```python
df3.to_csv("tuition_clean.csv", sep=',',index=False)
```
